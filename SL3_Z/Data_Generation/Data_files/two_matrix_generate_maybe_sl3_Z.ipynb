```python
import numpy as np
from collections import deque
import numpy as np
import time
import pandas as pd
```

```python
def matrix_to_tuple(matrix):
    return (matrix[0][0], matrix[0][1], matrix[0][2],
            matrix[1][0], matrix[1][1], matrix[1][2],
            matrix[2][0], matrix[2][1], matrix[2][2])

def tuple_to_matrix(my_tuple):
    return np.array([[my_tuple[0], my_tuple[1], my_tuple[2]], 
                     [my_tuple[3], my_tuple[4], my_tuple[5]], 
                     [my_tuple[6], my_tuple[7], my_tuple[8]]])

def apply_mat(mat, index):
    if index==0:
        return mat @ A
    elif index==1:
        return mat @ B
    assert(1==2)

# The commutator of the two elements in the group generated by the two matrices from Alex's paper has infinite index
A = np.array([[1, 1, 2], [0, 1, 1], [0, -3, -2]])
B = np.array([[-2, 0, -1], [-5, 1, -1], [3, 0, 1]])

I = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
```

```python
# adapted from ChatGPT
# visited contains keys of the tuples we've already visited and values of either 0 when A was the matrix that brought us to this state
# of 1 if B was the matrix that brought us to this state

visited = dict()

data = pd.DataFrame({
        'val1': [],
        'val2': [],
        'val3': [],
        'val4': [],
        'val5': [],
        'val6': [],
        'val7': [],
        'val8': [],
        'val9': [],
        'last_matrix': []
    })

# TODO we don't add A or B to the dataset

start_time = time.time()

def bfs(start, myDF):
    
    visited[matrix_to_tuple(start @ A)]=0
    visited[matrix_to_tuple(start @ B)]=1
    
    queue = deque([A, B])

    while queue:
        cur_mat = queue.popleft()

        example = matrix_to_tuple(cur_mat)
        # TODO maybe include the number of steps it took us to get here?

        if time.time() >= start_time + 500:
            print("Timed out")
            return myDF

        for neighbor in [0, 1]:
            new_mat = apply_mat(cur_mat, neighbor)
            new_tuple = matrix_to_tuple(new_mat)
            if not new_tuple in visited:
                visited[(new_tuple)]=neighbor
                cur_row_df = pd.DataFrame([[
                    example[0], example[1], example[2], 
                    example[4], example[5], example[6], 
                    example[6], example[7], example[8], 
                    neighbor
                    ]], columns=['val1', 'val2', 'val3', 'val4', 'val5', 'val6', 'val7', 'val8', 'val9', 'last_matrix'])
                myDF = pd.concat([cur_row_df, myDF])
                queue.append(new_mat)

data = bfs(I, data)
```
