```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import math
from PIL import Image
import os
```

```python
df = pd.read_csv("../Data_Generation/Data_files/bfs_heisenberg_data.csv")
```

```python
i=0

filtered_df = df[df['val1']==i]

plt.scatter(x=filtered_df['val2'], y=filtered_df['val3'], c=filtered_df['last_matrix'])
plt.xlabel("c")
plt.ylabel("b")
plt.title("When a=" + str(i))

plt.savefig("when_a_is_0.png")
```

```python
min_a = int(min(df['val1']))
max_a = int(max(df['val1']))

frames = []
fig = plt.figure(figsize=(8, 6))

for i in range(min_a, max_a+1):    
    filtered_df = df[df['val1']==i]

    plt.scatter(x=filtered_df['val2'], y=filtered_df['val3'], c=filtered_df['last_matrix'])
    plt.xlabel("c")
    plt.ylabel("b")
    plt.title("When a=" + str(i))

    file_name = f"a_{i}.png"
    plt.savefig(file_name)

    plt.clf()
    with Image.open(file_name) as img:
        frames.append(img.copy())
    os.remove(file_name)
```
