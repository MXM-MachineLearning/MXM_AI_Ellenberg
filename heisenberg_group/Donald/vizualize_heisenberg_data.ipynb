```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heisenberg_data.csv")
```

```python
test_df0 = df[df['last_matrix'] == 4]
test_df1 = df[df['last_matrix'] == 1]
test_df2 = df[df['last_matrix'] == 2]
test_df3 = df[df['last_matrix'] == 3]

plt.scatter(x=test_df0['val1'], y=test_df0['val2'], c="red", alpha=0.1)
plt.scatter(x=test_df1['val1'], y=test_df1['val2'], c="green", alpha=0.1)
plt.scatter(x=test_df2['val1'], y=test_df2['val2'], c="blue", alpha=0.1)
plt.scatter(x=test_df3['val1'], y=test_df3['val2'], c="black", alpha=0.1)


# adapted from https://www.geeksforgeeks.org/matplotlib-pyplot-legend-in-python/
plt.legend([4, 1, 2, 3])

plt.show()
```
