```python
import numpy as np
import torch
from torch import nn as nn
from torch import optim as optim
import time
```

```python
train_dataset = np.loadtxt('trainData.csv', delimiter=',')
X = train_dataset[:,0:2]
Y = train_dataset[:,2]

#device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device = torch.device("cpu")

X = torch.tensor(X, dtype=torch.float).to(device)
Y = torch.tensor(Y,dtype=torch.float).reshape(-1,1).to(device)

test_dataset = np.loadtxt('testData.csv', delimiter=',')
Xtest = test_dataset[:,0:2]
Ytest = test_dataset[:,2]

Xtest = torch.tensor(Xtest, dtype=torch.float).to(device)
Ytest = torch.tensor(Ytest,dtype=torch.float).reshape(-1,1).to(device)


class Euclid(nn.Module):
    def __init__(self):
        super(Euclid,self).__init__()
        self.layer1= nn.Linear(2,128)
        self.act1 = nn.ReLU()
        self.layer2= nn.Linear(128,64)
        self.act2 = nn.ReLU()
        self.layer3= nn.Linear(64,16)
        self.act3 = nn.ReLU()
        self.prediction = nn.Linear(16,1)
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self,x):        
        x = self.act1(self.layer1(x))
        x = self.act2(self.layer2(x))
        x = self.act3(self.layer3(x))
        x = torch.sigmoid(self.prediction(x))
        return x

    
model = Euclid().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
n_epochs = 100
batch_size = 10

start_time = time.time()
for epoch in range(n_epochs):
    for i in range(0, len(X), batch_size):
        Xbatch = X[i:i+batch_size]
        Ybatch = Y[i:i+batch_size]
        y_pred = model(Xbatch)
        loss = criterion(y_pred, Ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Finished epoch {epoch}, latest loss {loss}')
end_time = time.time()
total_time = end_time - start_time
print(f'Total training time: {total_time} seconds')

with torch.no_grad():
    y_pred = model(Xtest)
accuracy = (y_pred.round() == Ytest).float().mean()
print(f"Accuracy {accuracy}")
```
