```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

input_data = np.random.randint(-1000, 1000, size=(1000, 2))
desired_output = []

A = np.array([[1, 1], [0, 1]])
B = np.array([[1, -1], [0, 1]])
C = np.array([[1, 0], [1, 1]])
D = np.array([[1, 0], [-1, 1]])


for i in input_data:
    if i[0] >= 0 and i[1] >= 0 and i[0] >= i[1]:
        # this is matrix B
        desired_output.append(1)
    elif i[0] >= 0 and i[1] >= 0 and i[0] <= i[1]:
        # matrix D
        desired_output.append(3)
    elif i[0] <= 0 and i[1] >= 0 and i[1] >= -i[0]:
        # matrix C
        desired_output.append(2)
    elif i[0] <= 0 and i[1] >= 0 and i[1] <= -i[0]:
        # matrix A
        desired_output.append(0)
    elif i[0] >= 0 and i[1] <= 0 and i[0] >= -i[1]:
        # matrix A
        desired_output.append(0)
    elif i[0] >= 0 and i[1] <= 0 and i[0] <= -i[1]:
        # matrix C
        desired_output.append(2)
    elif i[0] <= 0 and i[1] <= 0 and i[1] <= i[0]:
        # matrix D
        desired_output.append(3)
    elif i[0] <= 0 and i[1] <= 0 and i[1] >= i[0]:
        # matrix B
        desired_output.append(1)
```

```python
df = pd.DataFrame(input_data)

df.columns = ["val1", 'val2']

df['matrix'] = desired_output
```

```python
plt.scatter(x=df['val1'], y=df['val2'], c=df["matrix"])
```

```python
df.to_csv('four_directions_cleaner.csv', index=False)
```
