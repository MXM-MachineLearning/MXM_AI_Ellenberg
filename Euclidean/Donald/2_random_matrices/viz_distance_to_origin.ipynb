```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
rand_seed = 42
```

```python
df = pd.read_csv(f"two_rand_matrices_seed_{rand_seed}.csv")
```

```python
# adapted from https://stackoverflow.com/questions/21702342/creating-a-new-column-based-on-if-elif-else-condition
def f(row):
    return math.sqrt(math.pow(row['val1'], 2) + math.pow(row['val2'], 2))

df['distance'] = df.apply(f, axis=1)
```

```python
def mySqrt(row):
    return math.sqrt(row['distance'])
```

```python
df['sqrt_dis'] = df.apply(mySqrt, axis=1)
```

```python
plt.hist(df['distance'], density=True, bins=200)

plt.show()
```

```python
plt.hist(df['sqrt_dis'], density=True, bins=200)

plt.show()
```
